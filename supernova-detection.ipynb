```python
chemin = '/kaggle/input/supernove-classification'
fichier = 'salt2_feature.csv'
chemin_fichier = os.path.join(chemin, fichier)

df = pd.read_csv(chemin_fichier , delimiter = ';')

matrix = df.values   #this will transforme the dataframe into numpy arrays 

df = pd.DataFrame(matrix)
print(df)

print("----------------------------------------------------------------------------------------------------------------")

encoder = OneHotEncoder(sparse_output = False)
print(df[6])
print("----------------------------------------------------------------------------------------------------------------")

ct = make_column_transformer(
    (encoder , [6]),
    remainder = 'drop'
)

ct.set_output(transform = 'pandas')

y = ct.fit_transform(df)

print(y.shape)
print(y)
```

```python
from sklearn.model_selection import train_test_split

data = np.array(data, dtype = np.float32)
print(data.shape)

X_train, X_test,y_train, y_test = train_test_split(data, y, random_state = 104, test_size=0.2, shuffle=True)

print(f"Training files: {len(X_train)}, label files: {len(y_train)}")
print(y_train.shape)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout, BatchNormalization, Reshape
from tensorflow.keras.utils import to_categorical
import tensorflow as tf

model = Sequential([
  Reshape((161, 7, 1), input_shape=(161, 7)),   #this reshapes the input to (161,7,1) before feeding it to the neural network
  BatchNormalization(),
  Conv2D(64, (3, 3), activation='relu', padding='same'), #means the ouput size is the same as the input size meaning padding zeros around the edges are needed
  BatchNormalization(),
  MaxPooling2D(pool_size=(2, 1)), #the window size is 2 by 1 and the stride is by default 1 meaning the height is reduced by 2
  Conv2D(32, (3, 3), activation='relu', padding='same'),
  BatchNormalization(),
  MaxPooling2D(pool_size=(2, 1)),
  Conv2D(16, (3, 3), activation='relu', padding='same'),
  BatchNormalization(),
  MaxPooling2D(pool_size=(2, 1)),
  Flatten(),
  Dense(256, activation='relu'),
  BatchNormalization(),
  Dropout(0.1),    #10% of the neurons are dropped randomly
  Dense(128, activation='relu'),
  BatchNormalization(),
  Dropout(0.1),
  Dense(64, activation='relu'),
  Dropout(0.1),
  Dense(3, activation='softmax')
  ])
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

from keras.callbacks import EarlyStopping

# Define EarlyStopping callback
early_stop = EarlyStopping(
    monitor='val_accuracy',     # Watch validation accuracy
    patience=5,             # Number of epochs to wait before stopping
    restore_best_weights=True  # Revert to best weights after stopping
)


model.fit(
    X_train, y_train,
    epochs=50,   #number if passes through the whole dataset
    validation_split=0.1, #reserves 10% of your training data to validation
    callbacks=[early_stop],
    batch_size=10
)
```

```python
print(X_test.shape)
print(y_test.shape)

test_loss, test_acc = model.evaluate(X_test, y_test) #test_loss = categorical cross entropy loss, test_acc = the accuracy of the model on test data.

print(f"\nTest accuracy: {test_acc:.4f}")

y_pred_probs = model.predict(X_test)

from sklearn.metrics import roc_auc_score #ROC-AUC (Receiver Operating Characteristic – Area Under Curve)

auc = roc_auc_score(y_test, y_pred_probs)
print("AUC Score:", auc)
```

```python
y_pred=model.predict(X_test)
y_pred=np.argmax(y_pred, axis=1)
y_test=np.argmax(y_test, axis=1)
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
